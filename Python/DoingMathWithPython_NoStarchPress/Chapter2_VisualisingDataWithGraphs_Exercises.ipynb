```python
#recreate np.linspace
def frange(start,final,increment):
    numbers=[]
    while start<final:
        numbers.append(start)
        start+=increment
    return numbers
```

```python
def draw_graph(x,y):
    plt.plot(x,y)
    plt.xlabel('x-coordinate')
    plt.ylabel('y-coordinate')
    plt.title('Projectile motion of a ball')
```

```python
#3: Enhanced Projectile Trajectory Comparison Program
import math
def draw_trajectory(u,theta):
    theta=math.radians(theta)
    g=9.8
    
    #Time of flight
    t_flight=2*u*math.sin(theta)/g
    print()
    print('Time of flight: {} seconds'.format(t_flight))
    #time intervals
    intervals=frange(0,t_flight,0.001)
    #x,y coords
    x,y=[],[]
    for t in intervals:
        x.append(u*math.cos(theta)*t)
        y.append(u*math.sin(theta)*t-0.5*g*t*t)

    print('Maximum vertical distance : {} m'.format(max(y)))
    #last value of x should be max x
    print('Maximum horizontal distance : {} m'.format(x[-1]))
    print()
    draw_graph(x,y)
```

```python
if __name__ == '__main__':
    try:
        trajectory_count=int(input('How many trajectories?'))
        for _ in range(trajectory_count):
            try:
                 u = float(input('Enter the initial velocity (m/s): '))
                 theta = float(input('Enter the angle of projection (degrees): '))
            except ValueError:
                 print('You entered an invalid input')
            else:
                 draw_trajectory(u, theta)
        plt.show()
    except ValueError:
        print('You entered an invalid input')
```

```python
#4 Visualising your  expenses
import matplotlib.pyplot as plt
def create_bar_chart(data,labels):
    #Number of bars
    num_bars=len(data)
    positions=range(1,num_bars+1)
    plt.barh(positions,data,align='center')
    plt.yticks(positions,labels)
    plt.xlabel('Steps')
    plt.ylabel('Day')
    plt.title('Number of steps walked')
    plt.grid()
    plt.show()
```

```python
#5: Exploring the Relationship Between the Fibonacci Sequence and the
#Golden Ratio
def fibo(n):
    if n==1:
        return [1]
    if n==2:
        return [1,1]
    #n>2
    a,b=1,1
    series=[a,b]
    for i in range(n):
        c=a+b
        series.append(c)
        a,b=b,c
    return series
```

```python
fibo(5)
```

```python
if __name__=="__main__":
    fib=fibo(100)
    fib_ratio=[fib[i+1]/fib[i] for i in range(len(fib)-1)]
    plt.plot(fib_ratio)
    plt.xlabel('Ratio')
    plt.ylabel('No')
    plt.title('Ratio between consecutive fibonacci numbers')
```
